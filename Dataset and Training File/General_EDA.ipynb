```python
# Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Data
df = pd.read_csv("/content/combined_dataset.csv")
df
```

# Exploratory Data Analysis

```python
# Showing data characteristics such as mean, median, etc
df.describe()
```

```python
# Examining the data type of each feature
df.info()
```

```python
# Checking missing value by calculating the percentage of missing value
missing_Val = 100 * (df.isna().sum() / len(df))
print(missing_Val)
```

```python
# Converting valid time data type into datetime
df['valid_time'] = pd.to_datetime(df['valid_time'])
# Sorting the valid time in ascending order
df = df.sort_values('valid_time')
```

```python
df = df.drop(columns=['pm1', 'pm10'], axis=1)
```

```python
# Showing Jakarta Time Plot
Jkt = df[df['city'] == 'Jakarta']
for col in ['pm2p5', 't2m', 'blh', 'bcaod550', 'omaod550']:
  sns.lineplot(data=Jkt, x='valid_time', y=col)
  plt.title(f"The Time Plot of {col} feature from Jakarta")
  plt.show()
```

```python
# Showing Helsinki Time Plot
Finland = df[df['city'] == 'Helsinki']
for col in ['pm2p5', 't2m', 'blh', 'bcaod550', 'omaod550']:
  sns.lineplot(data=Finland, x='valid_time', y=col)
  plt.title(f"The Time Plot of {col} feature from Helsinki")
  plt.show()
```

```python
# Showing Delhi Time Plot
delhi = df[df['city'] == 'Delhi']

for col in ['pm2p5', 't2m', 'blh', 'bcaod550', 'omaod550']:
  sns.lineplot(data=delhi, x='valid_time', y=col)
  plt.title(f"The Time Plot of {col} feature from Delhi City")
  plt.show()
```

```python
col = 'pm2p5'
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x=col, hue='city', element='step', common_norm=False, palette='tab10')
plt.xlabel(col)
plt.ylabel('Count')
plt.tight_layout()
plt.grid(True)

plt.savefig(f'{col}_distribution.png', dpi=300, bbox_inches='tight')
plt.close()

from google.colab import files
files.download('pm2p5_distribution.png')
```

```python
# Showing Box plot
for col in ['pm2p5']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='city', y=col, data=df)
    plt.tight_layout()
    plt.grid(True)

    # Save high-resolution image
    plt.savefig(f'{col}_boxplot.png', dpi=300, bbox_inches='tight')
    plt.close()

files.download('pm2p5_boxplot.png')
```

```python
# Correlation Heatmap
corr = df.drop(['valid_time', 'city'], axis=1).corr(method="pearson")
plt.figure(figsize=(12, 12))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap between Each Feature")
plt.show()
```
